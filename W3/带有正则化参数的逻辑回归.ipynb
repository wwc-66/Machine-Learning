```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost_regularized(x,y,w,b,lambda_):
    m = len(y)
    h = sigmoid(np.dot(x,w) + b)

    cross_entropy = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1-h))
    regularization = (lambda_/(2*m)) * np.sum(w**2)
    return cross_entropy + regularization

def gradient_decsent_regularized(x,y,w,b,lambda_,alpha,num_iters):
    m = len(y)
    cost_history = []

    for i in range(num_iters):
        h = sigmoid(np.dot(x,w) + b)
        errors = h - y

        dj_dw = (1/m) * np.dot(x.T, errors) + (lambda_/m) * w
        dj_db = (1/m) * np.sum(errors)

        w -= alpha * dj_dw
        b -= alpha * dj_db

        cost = compute_cost_regularized(x,y,w,b,lambda_)
        cost_history.append(cost)

    return w,b,cost_history
```
