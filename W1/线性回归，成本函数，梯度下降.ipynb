```python
import numpy as np
import matplotlib.pyplot as plt

#预测函数
def predict(x,w,b):
    return w * x + b

#成本函数
def compute_cost(x,y,w,b):
    m = len(y)
    total_cost = 0.0
    for i in range(m):
        total_cost += (predict(x[i],w,b)-y[i])**2
    return total_cost / (2*m)

#梯度下降
def gradient_descent(x,y,w,b,alpha,num_iters):
    m = len(y)
    cost_history = []
    for iter in range(num_iters):
        dj_dw = 0.0
        dj_db = 0.0
        for i in range(m):
            cost0 = predict(x[i],w,b) - y[i]
            dj_dw += x[i] * cost0
            dj_db += cost0

        w -= alpha * (dj_dw / m)
        b -= alpha * (dj_db / m)

        cost = compute_cost(x,y,w,b)
        cost_history.append(cost)
    return w,b,cost_history

#定义测试数据
X = np.array([1,2,3,4,5])
Y = np.array([1,1.8,3.2,3.7,5.1])

#运行梯度下降
w_initial = 0.0
b_initial = 0.0
alpha = 0.1
iterations = 50

print('开始梯度下降')
w_final,b_final,cost_history = gradient_descent(X,Y,w_initial,b_initial,alpha,iterations)
print(f'最终参数：w = {round(w_initial,3)}, b = {round(b_initial,3)}')

plt.figure(figuresize = (12,4))

plt.subplot
```
