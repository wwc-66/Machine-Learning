```python
import numpy as np

def softmax(z):
    exp_z = np.exp(z - np.max(z,axis = 1,keepdims = True))

    return exp_z / np.sum(exp_z,axis = 1,keepdims=True)

def test_softmax():
    logits = np.array([[2.0,1.0,0.1],
                       [1.0,2.0,0.1],
                       [0.5,1.0,2.5]])
    probabilities = softmax(logits)
    print(f'原始输入(logits) = {logits}')
    print(logits)
    print('/nsoftmax转换后的概率：')
    print(probabilities)
    print('/n每行概率和:',np.sum(probabilities))

    return probabilities

def cross_entropy_loss(probabilities, true_labels):
    m = true_labels.shape[0]
    loss = -np.sum(true_labels * np.log(probabilities+ 1e-8)) / m
    return loss
```
